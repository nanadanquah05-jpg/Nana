This notebook builds the Krackhardt Kite graph using NetworkX and visualizes it.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Create Nodes and Edges

The Krackhardt Kite has 10 nodes and a set of edges representing social connections.

```python
# Create undirected graph
G = nx.Graph()

# Add nodes
nodes = ['Andre', 'Beverly', 'Carol', 'Diane', 'Ed', 'Fernando', 'Garth', 'Heather', 'Ike', 'Jane']
G.add_nodes_from(nodes)

# Add edges based on the Krackhardt Kite structure
edges = [
    ('Andre', 'Beverly'),
    ('Andre', 'Carol'),
    ('Andre', 'Diane'),
    ('Andre', 'Fernando'),
    ('Beverly', 'Diane'),
    ('Beverly', 'Ed'),
    ('Beverly', 'Garth'),
    ('Carol', 'Diane'),
    ('Carol', 'Fernando'),
    ('Diane', 'Ed'),
    ('Diane', 'Fernando'),
    ('Diane', 'Garth'),
    ('Ed', 'Garth'),
    ('Fernando', 'Garth'),
    ('Fernando', 'Heather'),
    ('Garth', 'Heather'),
    ('Heather', 'Ike'),
    ('Ike', 'Jane')
]
G.add_edges_from(edges)

print(f'Nodes ({G.number_of_nodes()}): {list(G.nodes())}')
print(f'Edges ({G.number_of_edges()}): {list(G.edges())}')
```

## Visualize the Graph

```python
# Manual positions to approximate the kite layout
pos = {
    'Carol':    (2, 4),
    'Andre':    (0, 3),
    'Fernando': (4, 3),
    'Diane':    (2, 2),
    'Beverly':  (0, 1),
    'Garth':    (4, 1),
    'Ed':       (2, 0),
    'Heather':  (6, 2),
    'Ike':      (8, 2),
    'Jane':     (10, 2)
}

plt.figure(figsize=(12, 6))
nx.draw_networkx(
    G,
    pos=pos,
    with_labels=True,
    node_color='steelblue',
    node_size=1200,
    font_color='white',
    font_weight='bold',
    edge_color='gray',
    width=2
)
plt.title('Krackhardt Kite Graph', fontsize=14, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.savefig('krackhardt_kite.png', dpi=150)
plt.show()
```

## Basic Graph Metrics

```python
# Degree of each node
print('Node Degrees:')
for node, degree in sorted(G.degree(), key=lambda x: -x[1]):
    print(f'  {node}: {degree}')

print(f'\nDensity: {nx.density(G):.3f}')
print(f'Is connected: {nx.is_connected(G)}')
print(f'Average clustering coefficient: {nx.average_clustering(G):.3f}')
```

## Centrality Measures

```python
betweenness = nx.betweenness_centrality(G)
degree_centrality = nx.degree_centrality(G)

print('Betweenness Centrality (top 5):')
for node, val in sorted(betweenness.items(), key=lambda x: -x[1])[:5]:
    print(f'  {node}: {val:.3f}')

print('\nDegree Centrality (top 5):')
for node, val in sorted(degree_centrality.items(), key=lambda x: -x[1])[:5]:
    print(f'  {node}: {val:.3f}')
```
